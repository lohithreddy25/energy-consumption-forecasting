```python
df.drop(['total load forecast','forecast wind offshore eday ahead','generation hydro pumped storage aggregated'],
        axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['time'] = pd.to_datetime(df['time'], utc=True)
df['time'] = df['time'].dt.tz_localize(None)
```

```python
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month
df['day'] = df['time'].dt.day
df['hour'] = df['time'].dt.hour
df['day_of_week'] = df['time'].dt.dayofweek  # Monday=0, Sunday=6
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 24)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 24)
```

```python
df.drop('time',axis=1,inplace=True)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outlier_lower = Q1 - 1.5 * IQR
outlier_upper = Q3 + 1.5 * IQR

outliers = (df < outlier_lower) | (df > outlier_upper)
print("Outlier counts per column:\n", outliers.sum())
```

```python
df = df[~((df < outlier_lower) | (df > outlier_upper)).any(axis=1)]
print("Shape after removing outliers:", df.shape)
```

```python
unquie_value_arr = np.array(df.nunique())
print(unquie_value_arr)
```

```python
df.drop(['generation fossil coal-derived gas','generation fossil oil shale','generation fossil peat','generation geothermal','generation marine',
         'generation wind offshore'],axis=1,inplace=True)
```

```python
df.shape
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(24,16))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
df.drop(['forecast solar day ahead','forecast wind onshore day ahead','generation other renewable'],axis=1,inplace=True)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
X = df.drop('total load actual',axis=1)
y= df['total load actual']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

svm model

```python
from sklearn.svm import SVR
```

```python
svr_linear = SVR(kernel='linear')
svr_linear.fit(X_train, y_train)
y_pred_linear = svr_linear.predict(X_test)
r2_linear = r2_score(y_test, y_pred_linear)
mse_linear = mean_squared_error(y_test, y_pred_linear)
print(f"Linear Kernel R² Score: {r2_linear:.2f}")
print(f"Linear Kernel Mean Squared Error: {mse_linear:.2f}")
```

```python
svr_poly = SVR(kernel='poly', degree=3, C=1.0, epsilon=0.1)
svr_poly.fit(X_train, y_train)

y_pred_poly = svr_poly.predict(X_test)
r2_poly = r2_score(y_test, y_pred_poly)
print(f"Polynomial Kernel R² Score: {r2_poly:.2f}")
```

```python
svr_rbf = SVR(kernel='rbf', C=1.0, gamma=0.1, epsilon=0.1)
svr_rbf.fit(X_train, y_train)

y_pred_rbf = svr_rbf.predict(X_test)
r2_rbf = r2_score(y_test, y_pred_rbf)
mse_rbf = mean_squared_error(y_test, y_pred_rbf)
print(f"RBF Kernel R² Score: {r2_rbf:.2f}")
print(f"RBF Kernel Mean Squared Error: {mse_rbf:.2f}")
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
r2_lr=r2_score(y_test, y_pred_lr)

print(f"R² Score: {r2_lr:.2f}")
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
```

```python
y_pred = dt_model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)

rf_regressor.fit(X_train, y_train)

y_pred = rf_regressor.predict(X_test)

r2_rf = r2_score(y_test, y_pred)

print(f"R² Score: {r2_rf:.2f}")
```

```python

models = ['SVM_Linear', 'SVM_Polynomial','SVM_RBF', 'Linear Regression','RandomForestRegressor' , 'DecisionTreeRegressor']
r2_scores = [r2_linear, r2_poly,r2_rbf, r2_lr, r2_rf, r2]
plt.figure(figsize=(8, 6))
bars = plt.bar(models, r2_scores, color=['blue', 'orange', 'green', 'red', 'pink'])
plt.title('R-squared Scores for Different Models')
plt.xlabel('Model')
plt.ylabel('R-squared Score')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval/2, f'{yval:.2%}', ha='center', va='center', color='white')
plt.tight_layout()
plt.show()
```
